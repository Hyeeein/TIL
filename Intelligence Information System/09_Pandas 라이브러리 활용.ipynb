## Pandas 라이브러리 활용하기

#### 위키피디아 데이터 csv로 저장하기


1. pandas 라이브러리로 URL을 활용하여 데이터 가져오기

```python
import pandas as pd
```

```python
url = "https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84"
```

```python
df = pd.read_html(url)
df
```

2. df[0]을 활용하여 index 0에 해당하는 표 가져오기
    - 해당 페이지에 여러 개의 표가 존재 (10개)

```python
len(df)
```

```python
df[0]
```

3. 헤더를 열 이름(columns)으로, 나라 이름을 행 이름(index)로 지정

```python
df = pd.read_html(url, header=0, index_col=0)
df[0]
```

4. 슬라이싱을 활용하여 하계올림픽에 대한 데이터 추출

```python
a = df[0].iloc[:, :5]
a
```

5. Unnamed 부분 컬럼명 설정하기

```python
a.columns = ['하계 참가 횟수', '금', '은', '동', '계']
a
```

6. 금메달 기준으로 내림차순 정렬하기

```python
a = a.sort_values(by=['금'], ascending=False)
a
```

7. 해당 결과를 csv 파일로 저장하기

```python
a.to_csv('summer_medal.csv')
```
