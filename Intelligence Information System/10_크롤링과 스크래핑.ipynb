## 크롤링 (Crawling)

- 웹 페이지를 가져와서 데이터를 추출하는 행위

## 웹 기초

### 1. HTTP (Hyper Text Transfer Protocol)

* Hyper Text
    - 한 문서에서 다른 문서로 즉시 접근할 수 있는 텍스트
    - 하이퍼링크로 연결된 문서들을 클릭에 따라 자유롭게 이동
    
    
    
* HTTP: HTML로 작성되어 있는 **하이퍼 텍스트를 전송하기 위한 프로토콜**
    - 도메인 주소에서 http://



* 프로토콜: 약속, 규칙, 규약
    - A라는 지점에서 B라는 지점으로 데이터를 보낼 때 어떻게 보낼 것인지에 대한 규칙
    - **요청(request)와 응답(response)**
        - 서버(Server)와 클라이언트(Client)가 존재하여, 사용자(Client)가 웹 브라우저로 url을 통해 요청하면, 서버는 요청에 맞는 응답을 함
        - 요청은 GET과 url / 응답은 200과 HTML 문서

### 2. HTTPS (Hypertext Transfer Protocol Secure)



* 하이퍼 텍스트 전송 프로토콜 보안


* HTTP의 경우, 서버에서 클라이언트로 전송될 때, 정보가 암호화되지 않음 (데이터 도난의 위험)<br>
  -> **SSL(보안 소켓 계층)**을 사용하여 해결함<br>
  -> 서버와 브라우저 사이에 안전하게 암호화된 연결을 만들 수 있게 도와줌

### 3. URL (Uniform Resource Locator)


* 네트워크 상에서 **자원의 위치를 알려주는 주소**
    - 일반적으로 웹 브라우저에 주소 표시줄에 나오는 주소를 의미


* *URL이 https://www.naver.com인 경우?*
    - https라는 규약(프로토콜)으로 www.naver.com 이라는 주소의 실제 위치에 접속해서 정보를 가져옴
    
    
* http나 https 외에 ftp, mailto 같은 다른 규약도 존재

### 4. HTML (Hyper Text Markup Language)


* 웹 페이지를 위한 마크업 언어


* 마크업 언어 (Markup Language)
    - **웹 페이지를 작성하기 위한 문법들 중 한 종류**
    - 마크(Mark)로 둘러싸인 언어. 마크는 태크(tag)라고도 표현함
    - **문서의 구조**를 정의하기 위해 사용함 (문서 -> 문단 1/ 문단 2 -> 문장 1/ 문장 2/ 문장 3)


* cf) markdown
    - 마크업 언어의 일종으로, 존 그루버(John Gruber)와 아론 스워츠(Aaron Swartz)가 만들었음
    
    - 일반 텍스트 기반의 문서를 만들기 위한 경량 마크업 언어
        - 구조와 **서식**이 있는 문서 작성
        - **어떤 문자를 어떤 기능으로 인식하여 텍스트에 적용** (#을 이용해 제목 만드는 것 등)
        
    - 생활 속 마크다운 언어 (.md 확장자, 태그 기능(@가 붙은 문자열 자동인식))
    
    
* HTML의 구조 : 트리구조
    - 나뭇가지가 한 개에서 두 개로 각각 가지는 또 다른 나뭇가지로 늘어나면서 뻗어가는 구조
    - 나무와 비슷해서 트리 구조라는 이름이 붙음

## 웹 크롤링과 스크래핑


* 혼용되지만 미묘한 차이 존재


* 크롤링(Crawling)
    - **검색 엔진**이 웹 페이지의 정보를 수집하여 분류 및 색인화하여 DB화
    - 크롤러(crawler) 혹은 (bot)dl dnpq vpdlwlfmf qkdansgkdu epdlxj tnwlq
    - 검색 엔진에서 우리가 보는 것은 DB에 저장되어 있는 수집된 데이터를 보는 것 -> 빠르게 찾아서 보여줌
    
    
* 스크래핑
    - 웹 페이지의 정보를 수집하는 일련의 행위 -> 스크래핑은 크롤링을 포함하는 용어
    - 특정한 웹 페이에서 원하는 데이터 일부를 가져오는 것

## 스크래핑 주의사항

### 1. 크롤링과 법적 분쟁


* 크롤링(Crawling)

    - 웹 페이지를 그대로 가져와서 데이터를 추출하는 행위
    - 크롤링은 서비스 제공자가 의도한 방식이 아님

#### DB 침해 사례


1. 엠파스 열린검색 vs. 네이버 지식iN (2005)

    - 엠파스에서 검색하면 여러 검색엔진 결과를 가져옴
    - 네이버 지식iN에서 엠파스의 크롤링을 막는 것으로 끝
    
    
2. 리그베다위키 vs. 엔하위키미러

    - 리그베다위키의 승소: 엔하위키미러가 리그베다위키 사이트를 무단으로 미러링하여 무임승차를 통한 부정경쟁행위를 하고 있음 


3. 잡코리아 vs. 사람인

    - 잡코리아 승소 : 경쟁사(잡코리아)가 제공하는 채용정보를 허락 없이 크롤링해 자사 영업에 이용한 것은 부정경쟁행위
    
    * 잡코리아와 사람인의 법적 분쟁
    
        - 2008년 잡코리아는 사람인을 고소 -> 2017년에 법적분쟁 마무리
        - '부정경쟁행위'와 '데이터베이스 침해'라는 결과가 나옴
    
    * 사건 경위
    
        - 사람인이 잡코리아의 구인구직 정보를 크롤링
        - 만약, 사람인에 게시되지 않은 정보 발견하면 해당 회사에 연락해 구인 공고를 게시해도 되는지 허락을 구함. 허락할 경우 공고 게시
        - 잡코리아는 사람인 IP를 차단해버림. 사람인은 VPN을 이용해 또 다시 크롤링
        
    * 사람인의 주장
    
        - 웹 크롤링은 불법은 아님
        - 구인 회사의 허락을 받아 글을 올린 것
        - 구인 내용에 저작권이 있는 것이지, 사이트가 저작권을 가지고 있지 않음.
        
    * 잡코리아의 추가 대응
    
        - '데이터베이스권 침해'
            - 웹사이트는 저작권법상 데이터베이스에 해당
            - `잡코리아는 그 웹사이트의 제작 및 그 채용정보의 갱신/검증 또는 보충에 인적 또는 물적으로 투자했으므로 잡코리아 웹사이트에 대해 데이터베이스 제작자의 지위와 권리를 가짐`
            
     * 결국, 스크래핑 자체는 처벌받지 않음.
     * But, 데이터베이스권 침해 및 불공정 경쟁 => 수집하여 얻은 데이터에 추가적인 시간과 노력을 들였음. 따라서, 이런 자료를 상업적 용도로 사용할 땐 주의가 필요
    
    
4. 야놀자 vs. 여기어때
    - 야놀자 승소: 경쟁사인 야놀자의 제휴 숙박업소 목록 등을 영업 목적으로 수집 혐의

#### HiQ Labs vs. LinkedIn

* HiQ Labs에서 LinkedIn에 올라와 있는 사람들의 프로필을 수집하고, 이를 다른 정보들과 함께 가공 및 분석하여 판매

* 미국은 퍼블릭 정보(로그인이 필요 없는 암호화되어 있지 않은 정보)를 크롤링하는 것은 불법이 아니다.

#### 한국 vs. 미국

* 경쟁업체와의 부정경쟁행위? (상업적 이익/피해 유무)
    - 데이터베이스권 침해: 누군가가 시간과 비용을 들여 검증하고 가공한 데이터

* 공공데이터 (로그인 유무)
    - 웹 브라우저를 통해 확인할 수 있는 데이터
    
* 아직 관련 법과 인식이 정립되고 있는 시기

### 2. 스크래핑 주의사항

* 데이터를 스크래핑 하는 것 그 자체는 OK
    - `수집 과정`과 `수집 결과물의 상업적 활용 여부`에 따라 법적 문제가 발생할 수 있음
    
   
* 웹 서버에 무리가 가지 않게 수집 필요

    - 웹 서버에 지속적으로 데이터를 요청하는 행위는 DDos (분산 서비스 거부) 공격과 유사함. 웹 서버에 부하를 주는 경우 정상적인 서비스를 방해하는 행위가 됨
    
    - 너무 많은 요청 횟수가 있는 경우, 사용자를 차단할 수 있음
        - 웹 서버마다 허용 횟수 상이하고, time.sleep(10) 등을 활용함

- 상업적 이익을 발생시키는 경우 혹은 공공데이터가 아닌 데이터를 긁어오는 경우 NO
