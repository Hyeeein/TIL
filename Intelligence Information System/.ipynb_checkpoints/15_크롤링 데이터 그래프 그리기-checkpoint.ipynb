### 네이버 영화 데이터 수집 과정

1. 검색 대상 url 만들기
    - date 값을 변경하여 각 date 별 Top 50 데이터 가져오기
2. HTML 문서 파싱하기
    - 약 2분 정도 소요 (Top 50 X 100일 = 5000개)
3. 원하는 태그 값 추출하기
4. 추출한 값 리스트로 저장하기
5. 데이터 처리하기
6. 날짜 저장하기
7. 2일로 증가시키기
8. 100일로 증가시키기

### 진행상황 확인하는 법

1. tqdm 모듈 사용
    - Progress Bar(상태 진행바)를 지원
    - 현재의 진행 상황을 바(bar) 형태로 표현해줌
    - base 가상 환경에는 설치되어 있음
    - `from tqdm.notebook import tqdm`
    

2. 하단에 상태 진행바 생성 : 시간의 흐름에 따라 진행
    - <00:11의 의미
        - 코드가 완료되는데 총 11초 정도 걸릴 것이라는 것을 알려줌
    - 4.62it/s의 의미
        - it = iteration = 반복문 1회 도는 것
        - 1초에 반복문이 4.62번 수행됨

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm

# 7-1. 2일로 증가시키면서 periods를 2로 수정
# 8. 100일로 증가시키면서 periods를 100으로 수정
date_list = pd.date_range('2020-07-01', periods=100, freq='D')

# 7-2. 저장할 리스트 생성
movie_title = []
movie_point = []
movie_date = []

for today in tqdm(date_list):
    
    # 1. 검색 대상 url 만들기
    # 2. url을 따옴표를 이용하여 문자열로 만들어주기
    url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=" + (today.strftime('%Y%m%d'))
    html = urlopen(url)

    soup = BeautifulSoup(html.read(), "html.parser")
    
    # 3. 원하는 태그 값 추출하기
    titles = soup.find_all('div', {'class':'tit5'})
    points = soup.find_all('td', {'class':'point'})
    
    # 4. 추출한 값 리스트로 저장하기
    titles_text = []
    points_text = []
    
    # 6-1. 날짜 저장할 리스트
    date = []
    
    for t, p in zip(titles, points):
        
        # 5. 데이터 처리하기 (strip 함수)
        titles_text.append(t.text.strip(' \n'))
        points_text.append(float(p.text))
        
        # 6-2. 날짜 저장하기
        date.append(today)
    
    # 7-3. 리스트 + 리스트 = 두개가 합쳐진 하나의 큰 리스트 (concat)
    movie_title += titles_text
    movie_point += points_text
    movie_date += date
    
# 7-4. 슬라이싱해서 확인해보기
print(movie_title[:5])
print(movie_point[:5])
print(movie_date)
```

### 데이터 표로 만들기

```python
import numpy as np
import pandas as pd

frame = pd.DataFrame({'date':movie_date, 'title':movie_title, 'point':movie_point})
frame
```

### 피벗 테이블

* 피벗 테이블 (pivot table)
    - 커다란 표(데이터베이스, 스프레드시트 등)의 데이터를 요약하는 통계표
    - 요약 방식: 합계, 평균, 기타 통계 등 의미 있는 방식으로 묶음
    - 데이터 처리의 한 기법 / 유용한 정보에 집중할 수 있도록 하기 위해 통계를 정렬 또는 재정렬(피벗)하는 것

```python
# 피벗 테이블 연습

df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo", "bar", "bar", "bar", "bar"],
                  "B": ["one", "one", "one", "two", "two", "one", "one", "two", "two"],
                  "C": ["small", "large", "large", "small", "small", "large", "small", "small", "large"],
                  "D": [1, 2, 3, 3, 3, 4, 5, 6, 7],
                  "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
```

```python
# 행(index), 열(column), 값과 합계 지정하기
table = pd.pivot_table(df, values="D", index=['A', 'B'], columns=['C'], aggfunc=np.sum)
table
```

### 특정 영화 찾기

* `pandas.DataFrame.query` : 원하는 행을 필터링 혹은 선택하기
    1. 비교 연산자 (==, >, >=, <, <=, !=)
    2. in 연산자 (in, ==, not in, !=)
    3. 논리 연산자 (and, or, not)
    4. 외부 변수(또는 함수) 참조 연산
    5. 인덱스 검색
    6. 문자열 부분 검색 (str.contatins, str.startswith, str.endswith)

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm

date_list = pd.date_range('2020-07-01', periods=100, freq='D')

movie_title = []
movie_point = []
movie_date = []

for today in tqdm(date_list):
    url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=" + (today.strftime('%Y%m%d'))
    html = urlopen(url)

    soup = BeautifulSoup(html.read(), "html.parser")
    
    titles = soup.find_all('div', {'class':'tit5'})
    points = soup.find_all('td', {'class':'point'})
    
    titles_text = []
    points_text = []
    date = []
    
    for t, p in zip(titles, points):
        titles_text.append(t.text.strip(' \n'))
        points_text.append(float(p.text))
        date.append(today)
    
    movie_title += titles_text
    movie_point += points_text
    movie_date += date
    
print(movie_point)
```

### 특정 영화만 추려서 보기

```python
import pandas as pd

frame = pd.DataFrame({'date':movie_date, 'title':movie_title, 'point':movie_point})
frame
```

### query의 `== 연산자` 이용하기

- 1개의 경우에는 []를 생략할 수 있음
- 리스트로 값을 여러 개 받을 수 있음

```python
expr = "title == ['토이 스토리 3']"
search_movie = frame.query(expr)
print(search_movie)
```
