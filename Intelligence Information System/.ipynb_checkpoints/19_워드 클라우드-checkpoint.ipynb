```python
from collections import Counter

wc = WordCloud(background_color='white', max_words=200, font_path=r'C:\Windows\Fonts\batang.ttc')
count = Counter(ko_doc.split())
wc = wc.generate_from_frequencies(count)
count
```

```python
count.most_common(20)
```

```python
import matplotlib.pyplot as plt
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
ko_doc_noun = okt.nouns(ko_doc)
print(ko_doc_noun)
```

```python
from collections import Counter

count_noun = Counter(ko_doc_noun)

wc_noun = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\batang.ttc')
wc_noun = wc_noun.generate_from_frequencies(count_noun)
```

```python
count_noun.most_common(20)
```

```python
import matplotlib.pyplot as plt
plt.imshow(wc_noun, interpolation='bilinear')
plt.show()
```

```python
# 1글자로 이루어진 단어 제거하기

stopword_list = []

for noun_word in count_noun:
    if len(noun_word) == 1:
        stopword_list.append(noun_word)

stopword_list[:5]
```

```python
# 1글자인 단어를 모두 pop()

for stopword in stopword_list:
    count_noun.pop(stopword)
```

```python
wc_noun = wc_noun.generate_from_frequencies(count_noun)
```

```python
import matplotlib.pyplot as plt
plt.imshow(wc_noun, interpolation='bilinear')
plt.show()
```

```python
wc_noun = WordCloud(background_color='white',
                   max_words=2000, font_path=r'C:\Windows\Fonts\batang.ttc',
                   stopwords=stopword_list)
```

```python
import numpy as np
from PIL import Image

img_mask = np.array(Image.open('law.jpg'))

wc_img = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\batang.ttc', mask=img_mask)
wc_img = wc_img.generate_from_frequencies(count_noun)

import matplotlib.pyplot as plt
plt.imshow(wc_img, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
from urllib.request import urlopen
import urllib
import time
```

```python
url = "https://kin.naver.com/search/list.naver?query=%EB%B9%BC%EB%B9%BC%EB%A1%9C+%EB%8D%B0%EC%9D%B4"
html = urlopen(url)

soup = BeautifulSoup(html, "html.parser")
```

```python
tmp_list = []

for doc in docs:
    ans = doc.find_all("dd")[1]
    tmp_list.append(ans.text.strip())
    
print(tmp_list[:3])
```

```python
from tqdm.notebook import tqdm

tmp_list = []

for i in tqdm(range(1, 11)):
    url = "https://kin.naver.com/search/list.naver?query=%EB%B9%BC%EB%B9%BC%EB%A1%9C%20%EB%8D%B0%EC%9D%B4&page=" + str(i)
    html = urlopen(url)
    
    soup = BeautifulSoup(html, "html.parser")
    
    docs = soup.find("ul", {"class":"basic1"}).find_all("li")
    
    for doc in docs:
        ans = doc.find_all("dd")[1]
        tmp_list.append(ans.text.strip())
        
    time.sleep(0.5)
```

```python
from tqdm.notebook import tqdm

tmp_list = []

for i in tqdm(range(1, 1000)):
    url = "https://kin.naver.com/search/list.naver?query=%EB%B9%BC%EB%B9%BC%EB%A1%9C%20%EB%8D%B0%EC%9D%B4&page=" + str(i)
    html = urlopen(url)
    
    soup = BeautifulSoup(html, "html.parser")
    
    docs = soup.find("ul", {"class":"basic1"}).find_all("li")
    
    for doc in docs:
        ans = doc.find_all("dd")[1]
        tmp_list.append(ans.text.strip())
        
    time.sleep(0.5)

print(len(tmp_list))
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
ko_doc = '\n'.join(tmp_list)
print(ko_doc[:1000])
```

```python
print(len(ko_doc))
```

```python
import time

start = time.time()
ko_doc_noun = okt.nouns(ko_doc)
print(time.time() - start)
```

```python
from collections import Counter

count_noun = Counter(ko_doc_noun)
```

```python
count_noun.most_common(20)
```

```python
stopword_list = ['빼빼로데이', '빼빼로', '데이']

for word in count_noun:
    if len(word) == 1:
        stopword_list.append(word)

stopword_list[-5:]
```

```python
# stopword 단어를 모두 pop()

for stopword in stopword_list:
    if stopword in count_noun:
        count_noun.pop(stopword)
```

```python
from wordcloud import WordCloud

wc_noun = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\batang.ttc')
wc_noun = wc_noun.generate_from_frequencies(count_noun)

import matplotlib.pyplot as plt
plt.imshow(wc_noun, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
wc_noun = WordCloud(background_color='white',
                   max_words=2000, font_path=r'C:\Windows\Fonts\batang.ttc',
                   stopwords=stopword_list)
```

```python
import numpy as np
from PIL import Image

img_mask = np.array(Image.open('present.jpg'))

wc_img = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\batang.ttc', mask=img_mask)
wc_img = wc_img.generate_from_frequencies(count_noun)

import matplotlib.pyplot as plt
plt.imshow(wc_img, interpolation='bilinear')
plt.axis('off')
plt.show()
```
