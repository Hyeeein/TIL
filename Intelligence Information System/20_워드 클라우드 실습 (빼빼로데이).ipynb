## 지식iN 데이터 가져와서 그리기 (검색어: 빼빼로데이)

### 1. 필요한 라이브러리 import

```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
from urllib.request import urlopen
import urllib
import time
```

```python
# https://kin.naver.com/search/list.nhn?query=빼빼로%20데이&page=1

url = "https://kin.naver.com/search/list.naver?query=%EB%B9%BC%EB%B9%BC%EB%A1%9C+%EB%8D%B0%EC%9D%B4"
html = urlopen(url)

soup = BeautifulSoup(html, "html.parser")
```

### 2. 데이터 가져오기

* 태그값 확인



```python
docs = soup.find("ul", {"class":"basic1"}).find_all("li")

for doc in docs:
    ans = doc.find_all("dd")[1]
    print(ans)
```

```python
tmp_list = []

for doc in docs:
    ans = doc.find_all("dd")[1]
    tmp_list.append(ans.text.strip())
    
print(tmp_list[:3])
```

* page=str(i)로 만들어서 여러 페이지 정보 가져오기

```python
from tqdm.notebook import tqdm

tmp_list = []

for i in tqdm(range(1, 101)):
    url = "https://kin.naver.com/search/list.naver?query=%EB%B9%BC%EB%B9%BC%EB%A1%9C%20%EB%8D%B0%EC%9D%B4&page=" + str(i)
    html = urlopen(url)
    
    soup = BeautifulSoup(html, "html.parser")
    
    docs = soup.find("ul", {"class":"basic1"}).find_all("li")
    
    for doc in docs:
        ans = doc.find_all("dd")[1]
        tmp_list.append(ans.text.strip())
        
    time.sleep(0.5)
```

### 3. 데이터 전처리 

* '\n'.join(tmp_list) 이용하기


* Okt 이용하기

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
ko_doc = '\n'.join(tmp_list)
print(ko_doc[:1000])
```

* okt.nouns() 이용하여 명사만 추출하기

```python
print(len(ko_doc))
```

```python
import time

start = time.time()
ko_doc_noun = okt.nouns(ko_doc)
print(time.time() - start)
```

* collections의 Counter 이용하고, most_common() 확인

```python
from collections import Counter

count_noun = Counter(ko_doc_noun)
```

```python
count_noun.most_common(20)
```

* 불용어 처리

```python
stopword_list = ['빼빼로데이', '빼빼로', '데이']

for word in count_noun:
    if len(word) == 1:
        stopword_list.append(word)

stopword_list[-5:]
```

```python
# stopword 단어를 모두 pop()

for stopword in stopword_list:
    if stopword in count_noun:
        count_noun.pop(stopword)
```

### 4. 데이터 분석 및 시각화 (워드 클라우드)

```python
from wordcloud import WordCloud

wc_noun = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\batang.ttc')
wc_noun = wc_noun.generate_from_frequencies(count_noun)

import matplotlib.pyplot as plt
plt.imshow(wc_noun, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
wc_noun = WordCloud(background_color='white',
                   max_words=2000, font_path=r'C:\Windows\Fonts\batang.ttc',
                   stopwords=stopword_list)
```
