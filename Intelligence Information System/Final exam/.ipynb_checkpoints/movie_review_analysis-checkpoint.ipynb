*★ 리뷰와 평점 데이터 간의 갯수 차이 발생 ★*



- 위의 사진처럼 평점만 달린 경우, p 태그는 없고, 간혹 이름이 없는 경우도 있다.
- 따라서, cmt_info 클래스 전체를 가져와 \n으로 split 하고, 리스트의 길이를 비교하여 길이가 8인 것만 리뷰데이터와 평점 데이터에 추가했다.

```python
for i in tqdm(range(len(tmp_list))):
    tmp = tmp_list[i].text.split('\n')
    print(tmp)
```

```python
# 리뷰, 평점 내용 저장할 리스트
movie_reviews = []
ratings = []

# 리스트의 길이가 8이 되는 것만 새로운 리스트에 저장
for i in tqdm(range(len(tmp_list))):
    tmp = tmp_list[i].text.split('\n')
    # 결과 예시
    # ['떡국좋아', '2020. 10. 5. 22:40', '9', '웃다가 울다가. ..천만명이 선택 할 만 하다', '댓글 찬성하기', '10', '댓글 비추천하기', '5']
    
    if len(tmp) == 8:
        ratings.append(tmp[2])
        movie_reviews.append(tmp[3])
```

```python
print(len(movie_reviews), len(ratings))
```

```python
print(ratings)
```

### 데이터 전처리

- 리뷰가 없거나, '.', '..', '...'과 같은 리뷰는 제외
- 평점란에 리뷰가 들어간 경우도 리뷰와 평점 둘다 제외

```python
movie_reviews_new = []
ratings_new = []
```

```python
# 리뷰 내용이 없는 경우 제외하고 다시 저장
for i in tqdm(range(len(movie_reviews))):
    if movie_reviews[i] != ' ' or movie_reviews[i] != '.' or movie_reviews[i] != '..' or movie_reviews[i] != '...':
        movie_reviews_new.append(movie_reviews[i])
        ratings_new.append(ratings[i])
        
print(len(movie_reviews_new), len(ratings_new))
```

```python
movie_reviews_new_2 = []
ratings_new_2 = []
```

```python
# 평점란에 리뷰가 들어간 경우 리뷰와 평점 둘다 제외하고 다시 저장
for i in tqdm(range(len(ratings_new))):
    if len(ratings_new[i]) < 3:
        movie_reviews_new_2.append(movie_reviews_new[i])
        ratings_new_2.append(ratings_new[i])

print(len(movie_reviews_new_2), len(ratings_new_2))
```

```python
print(movie_reviews_new_2)
```

```python
print(ratings_new_2)
```

* 총 데이터를 전처리 한 후에 리뷰와 평점 모두 2120개의 데이터가 남았다.
* 리뷰는 모두 string 형태이고, 평점은 0~10점 사이의 string 형태로 구성되어있다.

### 2. 리뷰 데이터를 이용하여 워드 클라우드 그리기

* 영화 '7번방의 선물'에 대한 워드클라우드를 그렸다.
* 전체 리뷰 데이터는 총 2120개의 데이터로 그렸고, 긍정리뷰와 부정리뷰에 관해서는 아래에서 평점에 따라 나누고, 그에 맞게 전처리를 진행하였다.

#### (1) 전체 리뷰에 대한 워드 클라우드 그리기

- Okt 형태소 분석기를 활용해 명사만 추출해내는 전처리를 거쳤다.
- 그리고 most_common() 함수를 사용해 자주 등장하는 명사 20가지를 출력했고, 필요 없거나 의미 없는 단어의 경우 stopword로 판단하여 제거하였다.
- 워드클라우드의 mask는 수업 시간에도 활용했던 '선물'의 의미가 담긴 'present.jpg'를 활용하였다.

```python
# 전체 리뷰를 담은 리스트는 movie_reviews
# 한국어 형태소 분석기로 Okt 사용

from konlpy.tag import Okt
okt = Okt()

ko_doc = "\n".join(movie_reviews_new_2)
print(ko_doc)
```

```python
# okt.nouns() 이용하여 명사만 추출

start = time.time()
ko_doc_noun = okt.nouns(ko_doc)
print(time.time() - start)
```

```python
# collections의 Counter 이용하고, most_common() 확인

from collections import Counter

count_noun = Counter(ko_doc_noun)
count_noun.most_common(20)
```

* 불용어 처리는 추출된 명사 중 의미 없는 단어를 선택하여 직접 작성했다.
* 전체, 긍정, 부정 모두 직접 작성해서 보다 양질의 정보를 추출했다.

```python
# 불용어 처리

stopword_list = ['영화', '정말', '류승룡', '이', '진짜', '울', '것', '사람', '보고', '점']

for word in count_noun:
    if len(word) == 1:
        stopword_list.append(word)

stopword_list[-5:]
```

```python
# stopword 단어를 모두 pop()

for stopword in stopword_list:
    if stopword in count_noun:
        count_noun.pop(stopword)
```

```python
from wordcloud import WordCloud

wc_noun = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\batang.ttc')
wc_noun = wc_noun.generate_from_frequencies(count_noun)

import matplotlib.pyplot as plt
plt.imshow(wc_noun, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
wc_noun = WordCloud(background_color='white',
                   max_words=2000, font_path=r'C:\Windows\Fonts\batang.ttc',
                   stopwords=stopword_list)
```

```python
import numpy as np
from PIL import Image

img_mask = np.array(Image.open('present.jpg'))

wc_img = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\batang.ttc', mask=img_mask)
wc_img = wc_img.generate_from_frequencies(count_noun)

import matplotlib.pyplot as plt
plt.imshow(wc_img, interpolation='bilinear')
plt.axis('off')
plt.show()
```

**전체 리뷰에 대한 해석**

워드 클라우드 상에서 가장 잘 보이는 단어는 '눈물'과 '감동'이라는 단어이다.'7번방의 선물'이라는 영화에 대해 이야기를 하면 많은 사람이 보면서 눈물을 흘리고 감동을 받았다고 이야기했다.

친구들, 가족들과 이야기를 할 때에도 주인공이 딸 예승이를 지극정성으로 아끼고 그런 연기에서 웃음도 지을 수 있었고, 아버지가 많이 생각나는 감동적인 영화라는 말이 자주 나왔다.

결과적으로, 나와 내 친구들, 가족들이 느꼈던 것과 같이 많은 사람들이 이 영화를 보면서 눈물이 흐를 정도의 감동을 많이 느꼈음을 확인할 수 있다. 중간에 보이는 '연기', '최고', '아빠' 라는 단어를 보면서 딸을 향한 아버지의 사랑을 느낄 수 있던 영화였다는 점도 알 수 있었는데, 나만 그렇게 느꼈던 것이 아니라, 다른 사람들도 나와 비슷한 감정을 느꼈다는 점도 확인할 수 있었다.

### 리뷰데이터에 대해 ('review', 'rating') 형태로 데이터 수정하기

* 평점을 기준으로, 0-5점은 부정적인 리뷰, 6-10점은 긍정적인 리뷰로 생각했다.
* 중간층의 점수를 빼려고 했지만, 수집한 표본이 작으면 결과가 왜곡될 수 있다고 판단하여, 최대한 데이터들을 모두 살릴 수 있는 방안으로 진행했다.
* 그리고 이 과정을 진행하면서, 워드 클라우드에 사용할 수 있는 리뷰데이터 리스트, 나이브 베이즈 분류기에서 사용할 수 있는 튜플 형태의 리스트 두 가지 모두 생성했다.

```python
# 0~5는 부정 / 6~10은 긍정 리뷰라고 생각
review_ratings_positive = []
review_ratings_negative = []

review_ratings_positive_tuple = []
review_ratings_negative_tuple = []

for i in range(len(ratings_new_2)):
    if int(ratings_new_2[i]) >= 0 and int(ratings_new_2[i]) <= 5:
        review_ratings_negative.append(movie_reviews_new_2[i])
        review_ratings_negative_tuple.append((movie_reviews_new_2[i], 'neg'))
    else:
        review_ratings_positive.append(movie_reviews_new_2[i])
        review_ratings_positive_tuple.append((movie_reviews_new_2[i], 'pos'))
```

```python
print(len(review_ratings_positive), len(review_ratings_negative))
print(len(review_ratings_positive_tuple), len(review_ratings_negative_tuple))
```

* 긍정리뷰 데이터는 1857개, 부정리뷰 데이터는 263개이다.

```python
review_ratings = []

for i in range(len(review_ratings_positive_tuple)):
    review_ratings.append(review_ratings_positive_tuple[i])
    
for i in range(len(review_ratings_negative_tuple)):
    review_ratings.append(review_ratings_negative_tuple[i])
```

```python
print(len(review_ratings))
```

### 3. 리뷰데이터를 이용하여 나이브 베이즈 분류기 학습시키기

- 리뷰 데이터의 리뷰 텍스트와 점수를 이용하여 긍정/부정으로 분류하는 분류기 학습시키기
- Test 문장을 입력하여 분류기 결과 확인해보기

* 위에서 긍정 리뷰와 부정 리뷰로 나누는 과정에서 분류기에 사용할 수 있는 train 데이터셋을 만들었다.
    - 평점이 6-10점 사이는 **pos**로 분류했고, 평점이 0-5점 사이는 **neg**로 분류했다.
    - 하지만, 긍정 데이터와 부정 데이터의 비율 차이로 인해 **pos**가 조금 더 결과적으로 많이 나왔다. 이 문제를 해결하기 위해서 최대한 1:1 비율로 맞춰주는 것이 더 좋지만, 크롤링을 더 진행한다고 해서 부정 리뷰에 대한 데이터가 늘어나는 것이 아니라, 오히려 긍정 리뷰의 데이터가 늘어날 수 있기 때문에, 현재까지 다운 받은 데이터로만 학습을 시켜 사용했다. 이 점은 참고 부탁드립니다.

```python
from nltk.tokenize import word_tokenize
import nltk

from konlpy.tag import Okt
okt = Okt()

def pos_tokenize(raw_sent):
    pos_sent = []
    sent = okt.pos(raw_sent, norm=True, stem=True)
    
    for tup in sent:
        word, tag = tup[0], tup[1]    # tup: ('사과', 'Noun')
        word_tag = word + '/' + tag   # word_tag: '사과/Noun'
        pos_sent.append(word_tag)
        
    return ' '.join(pos_sent)

def make_word_dict(train, use_morph=False):
    all_words = set()
    
    for tup in train:
        sent, label = tup[0], tup[1]
        if use_morph: sent = pos_tokenize(sent)   # pos_tokenize 함수 추가
        words = word_tokenize(sent)
        for word in words:
            all_words.add(word)
            
    return all_words

def make_train_feats(train, all_words, use_morph=False):
    train_features = []
    
    for tup in train:
        sent, label = tup[0], tup[1]
        if use_morph: sent = pos_tokenize(sent)   # pos_tokenize 함수 추가
        words = word_tokenize(sent)
        tmp = {set_word: (set_word in words) for set_word in all_words}
        sent_tup = (tmp, label)
        train_features.append(sent_tup)
        
    return train_features
```
