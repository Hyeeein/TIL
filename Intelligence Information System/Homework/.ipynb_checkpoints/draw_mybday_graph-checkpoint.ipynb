## [과제 1] 매년 내 생일의 최고 기온 및 최저 기온 데이터 시각화하기

`인천대학교 경제학과 201900740 박혜인`

`생일: 2000년 8월 18일`

1. 그래프 옵션
    - 제목 : Hyein's Birthday Weather
    - 범례 : 자동으로 설정 (loc = 0)
    - 그래프 색 : 최고 기온 crimson / 최저 기온 dodgerblue
    - 그래프 선 모양 : dotted
    - 마커 : point ('.')

2. x축 이름과 y축 이름
    - x축 이름 : 횟수
    - y축 이름 : 기온

3. 다른 한글 폰트 사용
    - 사용한 폰트 : HYGothic-Medium

```python
import matplotlib.font_manager as fm

# 설치된 폰트 출력
font_list = [font.name for font in fm.fontManager.ttflist]
font_list
```

```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_weather.csv', 'r')
data = csv.reader(f)
next(data)

high = []       # 최고 기온 데이터를 저장할 빈 리스트 high 생성
low = []        # 최저 기온 데이터를 저장할 빈 리스트 low 생성

for row in data:

    # 최고 기온과 최저 기온 데이터 값이 존재한다면
    if row[-1] != '' and row[-2] != '':

        # 출생년도(2000년)이후 데이터 비교
        if int(row[0].split('-')[0]) >= 2000:
            
            # 8월 18일에 해당한다면
            if row[0].split('-')[1] == '08' and row[0].split('-')[2] == '18':
                high.append(float(row[-1]))     # high 리스트에 최고 기온 값 추가
                low.append(float(row[-2]))      # low 리스트에 최저 기온 값 추가

f.close()

# 그래프 사이즈 키우기
plt.figure(figsize=(12, 8))                

# 다른 한글 폰트 지정, 사이즈 15
plt.rc('font', family='HYGothic-Medium', size=15)   

plt.title('Hyein\'s Birthday Weather')     # 제목 지정
plt.xlabel('횟수')                         # x축 이름 지정
plt.ylabel('기온')                         # y축 이름 지정

# high 리스트에 저장된 값을 crimson 색 그래프로 그리기
plt.plot(high, color='crimson', label='최고 기온', linestyle='dotted', marker='.')

# low 리스트에 저장된 값을 dodgerblue 색 그래프로 그리기
plt.plot(low, color='dodgerblue', label='최저 기온', linestyle='dotted', marker='.')

plt.legend(loc=0)               # 범례 설정
plt.show()                      # 그래프 나타내기
```
