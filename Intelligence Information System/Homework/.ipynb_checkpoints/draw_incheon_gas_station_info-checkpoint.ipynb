### 2. 엑셀 파일 불러와서 합치기

* 반복문을 이용해 tmp_list에 저장 -> tmp_list[0] 값을 확인
* pd.concat을 활용해 여러 DataFrame 하나로 합치기

```python
import pandas as pd

tmp_list = []

for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_list.append(tmp)
    
tmp_list[0]
```

```python
stations = pd.concat(tmp_list)
stations.shape
```

```python
stations.info()
```

### 3. '휘발유', '경유'가 값으로 들어가는 '종류' 열 추가


(1) '휘발유'를 판매하는 주유소만 선택해 '상표', '상호', '주소', '셀프여부', '휘발유' 컬럼만 포함 <br>
(2) '휘발유' 열에서 값이 없는 것은 제외시킴 <br>
(3) '상호' 컬럼명을 '주유소'로, '휘발유' 컬럼명은 '가격'으로 바꿈
(4) '종류' 컬럼에 '휘발유' 값들로 추가함 <br>

-> (1)~(3)은 경유의 경우도 동일하게 진행 <br>

(5) 각 종류별로 나뉘어져 있는 '휘발유', '경유'를 pd.concat으로 합치기 <br>
(6) '군/구' 칼럼 추가하고, '상표' 컬럼을 인덱스로 설정

```python
# '휘발유'를 판매하는 주유소
stations_sel_1 = stations.loc[:, ['상표', '상호', '주소', '셀프여부', '휘발유']]
stations_sel_1 = stations_sel_1.query("휘발유 != '-'").copy()
stations_sel_1.columns = ['상표', '주유소', '주소', '셀프여부', '가격']
stations_sel_1['종류'] = '휘발유'
stations_sel_1
```

```python
# '경유'를 판매하는 주유소
stations_sel_2 = stations.loc[:, ['상표', '상호', '주소', '셀프여부', '경유']]
stations_sel_2 = stations_sel_2.query("경유 != '-'").copy()
stations_sel_2.columns = ['상표', '주유소', '주소', '셀프여부', '가격']
stations_sel_2['종류'] = '경유'
stations_sel_2
```

```python
# '휘발유', '경유' 판매하는 주유소 데이터 합치기
stations_sel = pd.concat([stations_sel_1, stations_sel_2])
stations_sel
```

```python
# 군/구 칼럼 추가

gu_list = []

for address in stations_sel['주소']:
    address_gu = address.split()[1]
    gu_list.append(address_gu)
    
stations_sel['군/구'] = gu_list
stations_sel
```

```python
stations_sel_name = stations_sel.set_index('상표')
stations_sel_name
```

### 4. 그래프 그리기 

- 그래프 그리기 위해 가격 데이터를 숫자로 변경
- matplotlib과 seaborn을 import하여 그래프 그리기

```python
# 가격을 숫자(int)으로 변경하기

price_list = []

for price in stations_sel['가격']:
    price_new = int(price)
    price_list.append(price_new)

stations_sel['가격'] = price_list
stations_sel
```

```python
stations_sel.info()
```

### 그래프 1

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")

plt.figure(figsize=(12, 8))
sns.boxplot(x="종류", y="가격", hue="셀프여부", data=stations_sel, palette="Set3")
plt.legend(title='셀프여부', loc=2)
plt.show()
```

`휘발유 값은 약 1600~1700원대 사이에 형성되어 있고, 경유 값은 약 1900원대 전후로 형성되어 있다. 따라서, 휘발유의 값이 경유보다 저렴하고, 기름의 종류에 따라 가격에 차이가 있음을 확인할 수 있다. 또, 셀프여부에 따라 살펴보면, 셀프 주유소의 가격이 셀프 주유소가 아닌 주유소보다 낮음을 확인할 수 있다.`

### 그래프 2

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.boxplot(x="상표", y="가격", hue="종류", data=stations_sel, palette="summer")
plt.legend(title='종류', loc=2)
plt.show()
```

`'자가상표'를 제외하고, 주유소 상표에 관계없이 휘발유는 약 1600원대로 가격이 약 1900원인 경유보다 전반적으로 저렴한 가격대를 형성하고 있음을 알 수 있다. 하지만 '자가상표'의 경우에는 다른 6가지 상표를 가진 주유소에 비해 휘발유 가격이 200~300원 정도 높게 형성되어 있고, 경유 또한 100원 정도 높게 형성되어 있다. 그럼에도 불구하고, '자가상표' 또한 휘발유가 경유보다 저렴하다.`

### 그래프 3

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.boxplot(x="군/구", y="가격", hue="종류", data=stations_sel, palette="summer")
plt.legend(title='종류', loc=2)
plt.show()
```

`휘발유의 가격은 웅진군을 제외하면 대체적으로 1600~1700원대 가격을 형성하고 있다. 경유의 경우에도, 웅진군을 제외하고 보통 1900원대 전후의 가격대를 형성하고 있다. 따라서, 휘발유가 경유보다 전반적으로 저렴함을 확인할 수 있다. 또, 웅진군과 중구의 경우에는 가격의 폭이 넓게 형성되어 있음을 확인할 수 있고, 반대로 계양구와 남구는 가격의 폭이 상대적으로 좁음을 확인할 수 있다.`
